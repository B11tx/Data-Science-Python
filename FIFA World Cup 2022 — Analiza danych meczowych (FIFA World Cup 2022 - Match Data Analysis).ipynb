https://www.kaggle.com/datasets/die9origephit/fifa-world-cup-2022-complete-dataset

Celem analizy jest zbadanie podstawowych statystyk dotyczących meczów Mistrzostw Świata 2022 w Katarze, w tym wyników bramkowych oraz trendów związanych z golami gospodarzy i gości.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("Fifa_world_cup_matches.csv")

print("Informacje o danych")
df.info()

print("\nStatystyki opisowe")
print(df.describe(include='all'))

print("\nBrakujące wartości")
print(df.isnull().sum())

df_clean = df.dropna()
print("\nLiczba wierszy po usunięciu braków:", len(df_clean))


print("\nAnalizy statystyczne")
print("Średnia liczba goli drużyny 1:", df_clean["number of goals team1"].mean())
print("Średnia liczba goli drużyny 2:", df_clean["number of goals team2"].mean())

max_goals = (df_clean["number of goals team1"] + df_clean["number of goals team2"]).max()
print("Maksymalna liczba goli w jednym meczu:", max_goals)

corr = df_clean["number of goals team1"].corr(df_clean["number of goals team2"])
print(f"Korelacja między golami drużyny 1 a drużyny 2: {corr:.2f}")

avg_goals = df_clean[["number of goals team1", "number of goals team2"]].mean()
avg_goals.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Średnia liczba goli — Drużyna 1 vs Drużyna 2 (Mundial 2022)")
plt.ylabel("Średnia liczba goli")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

Wnioski z analizy danych FIFA World Cup 2022

Na podstawie danych z Mundialu 2022 średnia liczba goli dla drużyn 1 wyniosła *1.58*,  
a dla drużyn 2 – *1.11*. 
Najwyższa liczba goli zdobyta w jednym meczu to *8*.  
Korelacja między golami obu drużyn *0.02* jest bardzo niska,  
co oznacza, że liczba bramek jednej drużyny nie wpływa znacząco  
na liczbę goli przeciwnika.  
Widać też, że średnio drużyny oznaczone jako „team1” (często grające jako gospodarze meczu)  
były nieco skuteczniejsze w ataku. 
Dane potwierdzają wyrównany, ale ofensywny charakter turnieju.
